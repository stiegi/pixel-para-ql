```python
import pandas as pd
import json
import re
```

```python
data = pd.read_excel("tableFromJS.xlsx")
data
```

```python
def add_type(row):
    parameter = row[0]
    if re.search('.*(?:cp|pu|cg).*',parameter) != None:
        return 'page'
    elif re.search('.*(?:ct|ck|is).*',parameter) != None:
        return 'action'
    elif re.search('.*(?:cs).*',parameter) != None:
        return 'session'
    elif re.search('.*(?:ba|ca|cb|co|cr|oi|ov|qn|st).*',parameter) != None:
        return 'ecommerce'
    elif re.search('.*(?:bw|mg|mi|mk|mt|mut|vol).*',parameter) != None:
        return 'media'
    elif re.search('.*(?:cc|mc).*',parameter) != None:
        return 'campaign'
    elif re.search('.*(?:cd|uc).*',parameter) != None:
        return 'urm'
    elif re.search('.*(?:ce).*',parameter) != None:
        return 'time'
    elif re.search('.*(?:fn|ft).*',parameter) != None:
        return 'form'
    else:
        return 'other'
    
data['type'] = data.apply(add_type, axis='columns')
data
```

```python
data.to_csv("parameterData.csv",sep="|", index=None)
```
